```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

df = pd.read_csv('IRIS.csv')

print(df.head())
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.dropna(inplace=True) # Drop rows with missing values
```

```python
missing_values = df.isnull().sum() # Checking again is there any missing data
print(missing_values)
```

```python
# Creating charts:
# Bar Chrats:
import matplotlib.pyplot as plt

plt.bar(df['species'], df['sepal_length'])
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Sepal Length by Species')
plt.show()
```

```python
# Pie Chart
plt.pie(df['species'].value_counts(), labels=df['species'].unique())
plt.title('Distribution of Species')
plt.show()
```

```python
# Line chart
plt.plot(df['species'], df['sepal_length'])
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Sepal Length comparison of  Species')
plt.show()
```

```python
# Histogram:
plt.hist(df['species'])
plt.xlabel('Species Names')
plt.ylabel('Count')
plt.title('Count of different species')
plt.show()
```

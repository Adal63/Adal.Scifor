```python
# Q1: Implement Linear or Multiple linera regression

# Linear Regression:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
iris = load_iris()
x = iris.data[:,0].reshape(-1, 1)
y = iris.data[:,2].reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_predict = model.predict(y_test)
```

```python
MSE = mean_squared_error(y_test,y_predict)
R2_SCORE = r2_score(y_test,y_predict)
```

```python
print(f"Mean squared Error:{MSE}")
print(f"R2 Score:{R2_SCORE}")
```

```python
# Q2: Decision tree with hyper parameter tuning.

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
iris = load_iris()
X,y = iris.data,iris.target
```

```python
DTClassifier = DecisionTreeClassifier()
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(estimator=DTClassifier,param_grid = param_grid,cv=5,n_jobs=-1,verbose=1)
grid_search.fit(X, y)
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
# Q3: Implementation of Naive Baye's Classifier.

from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
iris = load_iris()
x,y = iris.data,iris.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
y_predict = gnb.predict(x_test)
```

```python
Accuracy = accuracy_score(y_test,y_predict)
Confusion_matrix = confusion_matrix(y_test,y_predict)
```

```python
print(f"Accuracy:{Accuracy}")
print(f"Accuracy:{Accuracy}")
```

```python
# Q4: Implementation of Support Vector Machine:

from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
iris = load_iris()
x,y = iris.data,iris.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
svm = SVC(kernel='linear')
```

```python
svm.fit(x_train,y_train)
y_predict = svm.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_predict))
print("Confusion matrix:",confusion_matrix(y_test,y_predict))
print("Classification Report:",classification_report(y_test,y_predict))
```

```python
# Q6: Handling missing values explaination with implementation.

# Handling missing values can be done using various methods such as dropping, filling with mean/median/mode, or using more complex imputation techniques.

import pandas as pd
from sklearn.impute import SimpleImputer

# Sample data
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, 4, np.nan],
        'C': [1, np.nan, np.nan, 4, 5]}

df = pd.DataFrame(data)

# Impute with mean
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("Original Data:")
print(df)
print("\nImputed Data:")
print(df_imputed)
```
